```python
import numpy as np
import pandas as pd

df = pd.read_csv('C:/Users/Admin/Desktop/temp/Numpy/terrorismData.csv',encoding='utf-8')
# year
year = len(set(df['Year']))

df = df[df['Country'] == 'India']
df['Casuality'] = df['Killed'] + df['Wounded']

jk = df[df['State'] == 'Jammu and Kashmir']
# | -> set each bit to 1 if one of them is 1 (all values are true)
rcc = df[(df['State'] == 'Chhattisgarh') | (df['State'] == 'Jharkhand') | (df['State'] == 'Andhra Pradesh') | (df['State'] == 'Odisha')]

jkc = int(np.sum(jk['Casuality']))
rccc = int(np.sum(rcc['Casuality']))

print(jkc//year,rccc//year)
```
